# Hierarchical Clustering 

## Introduction
Hierarchical clustering is a type of clustering algorithm that builds a hierarchy of clusters. It starts with each data point as a separate cluster and then merges the closest pairs of clusters until all points belong to just one big cluster. In this tutorial, we'll learn how to perform hierarchical clustering using Python and visualize the results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage 
```

```python
# Generate sample data
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

### Perform Hierarchical Clustering
We'll use the AgglomerativeClustering class from scikit-learn to perform hierarchical clustering:

```python
# Perform hierarchical clustering
model = AgglomerativeClustering()
model.fit(X)
```

### Visualize Dendrogram
To visualize the hierarchy of clusters, we can plot a dendrogram using scipy.cluster.hierarchy.dendrogram:

```python
# Plot dendrogram
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and plot dendrogram
    linkage_matrix = linkage(X, method='ward')  # Change the method if needed
    dendrogram(linkage_matrix, **kwargs)

plt.figure(figsize=(12, 6))
plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, labels=model.labels_)
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

### Visualize Clusters
Finally, let's visualize the clusters on a scatter plot:

```python
# Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='viridis')
plt.title('Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.show()
```
